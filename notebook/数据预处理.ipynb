```python
exclude_cols = ['国家', '州', '城市', '邮政编码', '经纬度组合', '纬度', '经度', '客户ID', '计数', '流失原因', '流失标签', '流失评分', '客户终身价值']
service_cols = [
    '电话服务', '多条线路', '互联网服务', '网络安全', 
    '网络备份', '设备保护', '技术支持', '流媒体电视', 
    '流媒体电影', '合同类型', '无纸化账单', '支付方式'
]

# 2. 打印每个类别的取值可能性
print("--- 各服务字段的取值分布 ---")
for col in service_cols:
    unique_vals = df[col].unique()
    print(f"【{col}】: {unique_vals}")
```

```python
plt.figure(figsize=(20, 15))

for i, col in enumerate(service_cols):
    ax = plt.subplot(4, 3, i + 1)
    
    crosstab = pd.crosstab(df[col], df['流失标签'])
    
    crosstab_prop = crosstab.div(crosstab.sum(1), axis=0)

    crosstab_prop.plot(kind='bar', stacked=True, ax=ax, color=['#66b3ff', '#ff9999'])
    
    ax.set_title(f'{col} vs 流失', fontsize=12, fontweight='bold')
    ax.set_xlabel('')
    ax.set_ylabel('比例')
    ax.legend(title='流失', loc='upper right', frameon=True, fontsize='small')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy import stats

# 1. 定义数值型列
numeric_cols = ["在网时长（月）", "月费用", "流失值", "流失评分", "客户终身价值"]

# 确保总费用是数值型 (有时候可能会因为有空格被当成字符串)
df['总费用'] = pd.to_numeric(df['总费用'], errors='coerce')
# 如果有因为转换产生的空值（通常是新用户），填充为0或均值，这里暂时填0
df['总费用'] = df['总费用'].fillna(0)

# 2. 定义检测函数
def detect_outliers(df, col):
    data = df[col]
    
    # --- 方法 A: Z-score ---
    # 阈值通常设为 3 (即偏离均值 3 个标准差)
    z_scores = np.abs(stats.zscore(data))
    z_outliers = np.where(z_scores > 3)[0]
    
    # --- 方法 B: IQR (四分位距) ---
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    iqr_outliers = df[(data < lower_bound) | (data > upper_bound)].index
    
    return len(z_outliers), len(iqr_outliers), lower_bound, upper_bound

# 3. 执行检测并打印报告
print(f"{'列名':<15} | {'Z-score异常数':<12} | {'IQR异常数':<10} | {'IQR下界':<10} | {'IQR上界':<10}")
print("-" * 75)

for col in numeric_cols:
    z_num, iqr_num, lb, ub = detect_outliers(df, col)
    print(f"{col:<15} | {z_num:<12} | {iqr_num:<10} | {lb:<10.2f} | {ub:<10.2f}")
```

## 数据变换

```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


scale_cols = ['在网时长（月）', '月费用', '总费用', '客户终身价值']

scaler = StandardScaler()

df_norm = df.copy()
df_norm[scale_cols] = scaler.fit_transform(df[scale_cols])

print("--- 标准化后的统计特性 (均值应接近0，标准差应接近1) ---")
desc = df_norm[scale_cols].describe().loc[['mean', 'std', 'min', 'max']]
print(desc.round(2))

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.histplot(df['客户终身价值'], kde=True, ax=axes[0], color='skyblue')
# axes[0].set_title('原始数据: 客户终身价值')
axes[0].set_xlabel('原始金额')

sns.histplot(df_norm['客户终身价值'], kde=True, ax=axes[1], color='green')
# axes[1].set_title('StandardScaler 后: 客户终身价值')
axes[1].set_xlabel('Z-score (标准差倍数)')

plt.tight_layout()
plt.show()
```
